***01. 「パタトクカシーー」***<br>
「パタトクカシーー」という文字列の1,3,5,7文字目を取り出して連結した文字列を得よ．

```julia
#1-01
S="パタトクカシーー"

index_list=[]
n=firstindex(S)

while n ≤ lastindex(S)
    push!(index_list,n)
    n=nextind(S,n)
end
res=""
for i in index_list[1:2:end]
    res*=S[i]
end

print(res)
```

***03. 円周率***<br>
“Now I need a drink, alcoholic of course, after the heavy lectures involving quantum mechanics.”という文を単語に分解し，各単語の（アルファベットの）文字数を先頭から出現順に並べたリストを作成せよ．

```julia
#1-03
S="Now I need a drink, alcoholic of course, after the heavy lectures involving quantum mechanics."
count=0
res=[]
for s in S
    if s ==' ' || s == ',' || s == '.'
        if count>0
            push!(res,count)
        end
        count=0
    else
        count+=1
    end
end
print(res)
```

***04. 元素記号***<br>
“Hi He Lied Because Boron Could Not Oxidize Fluorine. New Nations Might Also Sign Peace Security Clause. Arthur King Can.”という文を単語に分解し，1, 5, 6, 7, 8, 9, 15, 16, 19番目の単語は先頭の1文字，それ以外の単語は先頭の2文字を取り出し，取り出した文字列から単語の位置（先頭から何番目の単語か）への連想配列（辞書型もしくはマップ型）を作成せよ．

```julia
#1-04
S="Hi He Lied Because Boron Could Not Oxidize Fluorine. New Nations Might Also Sign Peace Security Clause. Arthur King Can."
word_list=split(S,r"[,\. ] ?", keepempty=false)
#↑この一行は,https://qiita.com/Rshimura/items/4389d164840589a14ec6　より
first_char_list=[1, 5, 6, 7, 8, 9, 15, 16, 19]
res=Dict()

#先頭1文字
for i in first_char_list
    res[i]=word_list[i][1]
end

j=1
#先頭2文字
for i in 1:length(word_list)
    if j ≤ length(first_char_list) && i == first_char_list[j]
        j+=1
        continue
    end
    res[i]=word_list[i][1:2]
end

res
#ソートができません
```

***05. n-gram***<br>
与えられたシーケンス（文字列やリストなど）からn-gramを作る関数を作成せよ．この関数を用い，”I am an NLPer”という文から単語bi-gram，文字bi-gramを得よ．

```julia
#1-05
S="I am an NLPer"

#単語bi-gram
word_list=split(S,r"[,\. ] ?", keepempty=false)
res=[]
for i in 1:3
    push!(res,(word_list[i],word_list[i+1]))
end
println(res)

#文字bi-gram
res=[]
for i in 1:length(S)-1
    push!(res,(S[i],S[i+1]))
end
println(res)


# [別解(というかこちらの方が題意に沿っている)] https://qiita.com/Rshimura/items/4389d164840589a14ec6を参照した

function gen_ngram(c,n)
    [c[i:(i+n-1)] for i in 1:length(c)-n+1]
end

println(gen_ngram(S,2))
println(gen_ngram(split(S),2))
```

***06. 集合***<br>
“paraparaparadise”と”paragraph”に含まれる文字bi-gramの集合を，それぞれ, XとYとして求め，XとYの和集合，積集合，差集合を求めよ．さらに，’se’というbi-gramがXおよびYに含まれるかどうかを調べよ．

```julia
#1-06
S1="paraparaparadise"
S2="paragraph"

X=Set(gen_ngram(S1,2))
Y=Set(gen_ngram(S2,2))
print("和集合 : ")
println(union(X,Y))
print("積集合 : ")
println(intersect(X,Y))
print("差集合 : ")
println(setdiff(X,Y),"\t",setdiff(Y,X))
println("se in X : ","se" in X)
println("se in Y : ","se" in Y)
```

***08. 暗号文***<br>
与えられた文字列の各文字を，以下の仕様で変換する関数cipherを実装せよ．

英小文字ならば(219 - 文字コード)の文字に置換
その他の文字はそのまま出力
この関数を用い，英語のメッセージを暗号化・復号化せよ.

```julia
function cipher(S)
    res=""
    for s in S
        if 'a' ≤ s ≤ 'z'
            s=Char(219-codepoint(s))
        end
        res*=s
    end
    res #return res
end
print(cipher("Kin-iro Mosaic")) # [参照]http://www.kinmosa.com/
```

***09. Typoglycemia***<br>
スペースで区切られた単語列に対して，各単語の先頭と末尾の文字は残し，それ以外の文字の順序をランダムに並び替えるプログラムを作成せよ．ただし，長さが４以下の単語は並び替えないこととする．適当な英語の文（例えば”I couldn’t believe that I could actually understand what I was reading : the phenomenal power of the human mind .”）を与え，その実行結果を確認せよ．

```julia
#1-09
using Random

function randomize_string(word)
    list=[]
    for s in word
        push!(list,s)
    end
    join(shuffle(list))
end
function Typoglycemia(S)
    res_S=[s for s in S]
    word_list=split(S,r"[,\. ] ?", keepempty=false)
    str_tmp=""
    for word in word_list
        if length(word) ≤ 3
            str_tmp*=word
            continue
        end
        w_tmp=word[2:end-1]
        str_tmp*=word[1]*randomize_string(w_tmp)*word[end]
    end
    j=1
    for i in 1:length(S)
        if S[i]!=' ' && S[i]!=',' && S[i]!='.'
            res_S[i]=str_tmp[j]
            j+=1
        end
    end
    join(res_S)
end

S="I couldn't believe that I could actually understand what I was reading : the phenomenal power of the human mind ."
Typoglycemia(S)
```
